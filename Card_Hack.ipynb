```python
X = df.drop('Profit in $',axis=1)
y = df["Profit in $"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_Scaled,y,test_size=0.25,random_state=255)
```

```python
import statsmodels.api as sm

Xc = sm.add_constant(X_Scaled)
lrm = sm.OLS(y,Xc).fit()
lrm.summary()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train,y_train)

pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,roc_auc_score,mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,pred))

r2 = r2_score(y_test,pred)
```

```python
r2
```

```python
rmse
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

rf.fit(x_train,y_train)

pred1 = rf.predict(x_test)
```

```python
rmse_1 = np.sqrt(mean_squared_error(y_test,pred1))

r2_1= r2_score(y_test,pred1)
```

```python
r2_1
```

```python
from sklearn.ensemble import AdaBoostRegressor

ab = AdaBoostRegressor()

ab.fit(x_train,y_train)

pred2 = ab.predict(x_test)
```

```python
rmse_2 = np.sqrt(mean_squared_error(y_test,pred2))

r2_2= r2_score(y_test,pred2)
```

```python
r2_2
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()

dt.fit(x_train,y_train)

pred3 = dt.predict(x_test)
```

```python
rmse_3 = np.sqrt(mean_squared_error(y_test,pred3))

r2_3= r2_score(y_test,pred3)
```
